```python
import numpy as np
import matplotlib.pyplot as plt #library for visulation عشان أظهر الداتا اللي عندي ف رسومات
```

```python
x=np.arange(0,10)
y=np.sin(x)
y
```

```python
plt.plot(x,y,color="yellow" ,linestyle="dashed", marker=">" ,markerfacecolor="red",markersize=4,linewidth=3)
plt.xlabel("x")
plt.ylabel("y=sin(x)")
plt.title("show data ", color="red")
plt.show()
```

```python
z=np.cos(x)
plt.plot(x,z,'b:o' ,x,y,'y-->')#color line marker
plt.legend(["cos","sin"] ,loc="best")
plt.show()
```

```python
x=dt.iloc[:,:2]
x
```

```python
from sklearn.cluster import KMeans
```

```python
km=KMeans(n_clusters=3)
```

```python
cluster_labels=km.fit_predict(x)
cluster_labels
```

```python
dt["cluster_labels"]=cluster_labels
dt
```

```python
km.cluster_centers_
```

```python
km.labels_
```

```python
km.n_iter_
```

```python
group0=dt[dt.cluster_labels==0]
group0
```

```python
group1=dt[dt.cluster_labels==1]
group2=dt[dt.cluster_labels==2]
```

```python
plt.scatter(group0.sepal_length,group0.sepal_width,label="cluster=0")
plt.scatter(group1.sepal_length,group1.sepal_width,label="cluster=1")
plt.scatter(group2.sepal_length,group2.sepal_width,label="cluster=2")
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],label="centeroid",marker="*",color="red")
plt.legend()
plt.show()
```
